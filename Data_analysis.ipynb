```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import numpy as np
```

```python
df= pd.read_csv('Tuberculosis dataset.csv')
```

```python
grouped = df.groupby('Gender')['HIV_Prevalence'].sum()
fig,ax=plt.subplots(figsize=(12,5))
ax.pie(grouped.values, labels=grouped.index, autopct='%1.1f%%', startangle=90)
ax.axis('equal')
plt.title('HIV Prevalence Distribution by Gender')
plt.show()

```

```python
pivoted=df.pivot_table(columns='Age_Group', values='HIV_Prevalence', aggfunc='mean')
pivoted.T.plot(kind='bar')
plt.title('HIV Prevalence by Age Group')
plt.xlabel('Age Group')
plt.ylabel('HIV Prevalence')
plt.xticks(rotation=0)
plt.show()
```

```python
pivoted_df = df.pivot_table(columns='Country',values='HIV_Prevalence',aggfunc='mean')
pivoted_df = pivoted_df.T
Top_n = pivoted_df.nlargest(10, 'HIV_Prevalence')
Top_n.sort_values(by='HIV_Prevalence', ascending=False, inplace=True)
fig, ax = plt.subplots(figsize=(12, 5))
sns.barplot(x=Top_n.index, y='HIV_Prevalence', data=Top_n.reset_index(), ax=ax)
plt.title('Top 10 Countries with Highest HIV Prevalence')
plt.xlabel('Country')
plt.ylabel('HIV Prevalence')
plt.show()

```

```python
pivoted_year = df.pivot_table(columns='Year',values='HIV_Prevalence',aggfunc='sum')
rate_pivoted_year = df.pivot_table(columns='Year',values='HIV_CoInfection_Rate',aggfunc='sum')
pivoted_transposed = pivoted_year.T
rate_pivoted_transposed = rate_pivoted_year.T

```

```python

fig, axs = plt.subplots(1, 2, figsize=(12, 5))
sns.lineplot(rate_pivoted_transposed, ax=axs[0])
sns.lineplot(pivoted_transposed, ax=axs[1])
axs[1].set_title('HIV Prevalence Over Time')
axs[1].set_xlabel('Year')
axs[1].set_ylabel('HIV Prevalence')
axs[0].set_title('HIV Coinfection Rate Over Time')
axs[0].set_xlabel('Year')
axs[0].set_ylabel('HIV Coinfection Rate')
plt.show()

```

```python
fig,ax=plt.subplots(figsize=(12,5))
sns.scatterplot(df, x='HIV_Prevalence', y='TB_Incidence_Rate', hue='Gender', ax=ax)
ax.set_title('HIV Prevalence vs TB Incidence Rate')
ax.set_xlabel('HIV Prevalence')
ax.set_ylabel('TB Incidence Rate')
plt.show()
```

```python
sns.kdeplot(data=df, x='TB_Incidence_Rate', hue='Gender', palette='bright',fill=True)
plt.title('TB Incidence Rate by Gender')
plt.xlabel('TB Incidence Rate')
plt.ylabel('Density')
plt.show()
```

```python
corr = df.corr(numeric_only=True)
```

```python
plt.figure(figsize=(12, 5))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=1, fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
TB_year = df.pivot_table(index='Gender',columns='Year',values='TB_Incidence_Rate',aggfunc='sum')
TB_year = TB_year.T
```

```python
fig,ax=plt.subplots(figsize=(12,5))
sns.lineplot(TB_year, ax=ax)
ax.set_title('TB Incidence Rate Over Time')
ax.set_xlabel('Year')
ax.set_ylabel('TB Incidence Rate')
```

```python
grouped=df.groupby('Gender')['Population'].max()
```

```python
fig,ax=plt.subplots(figsize=(12,5))
ax.pie(grouped.values, labels=grouped.index, autopct='%1.1f%%', startangle=90)
ax.axis('equal')
plt.title('Population Distribution by Gender')
plt.show()
```

```python
pivoted=df.pivot_table(columns='Age_Group', values='Alcohol_Consumption', aggfunc='count')
T_transposed = pivoted.T

```

```python
fig,ax=plt.subplots(figsize=(12,5))
sns.barplot(T_transposed, x=T_transposed.index, y='Alcohol_Consumption', ax=ax)
ax.set_title('Alcohol Consumption by Age Group')
ax.set_xlabel('Age Group')
ax.set_ylabel('Alcohol Consumption')
plt.show()
fig.show()
```
